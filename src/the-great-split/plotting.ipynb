```python
import numpy as np
```

```python
prediction_dat = np.loadtxt('./final_exam_predictions.dat', delimiter=',')
```

```python
y_true = prediction_dat[:, 0]
```

```python
preds_dict = {key: [] for key in set(y_true)}
for true_val, pred_val in zip(y_true, prediction_dat[:, 1]):
    preds_dict[true_val].append(pred_val)
```

```python
true_plot = []
pred_plot = []
uncert_plot = []
for key, val in preds_dict.items():
    true_plot.append(key)
    pred_plot.append(np.mean(val))
    uncert_plot.append(np.std(val))
    
```

```python
import matplotlib.pyplot as plt
```

```python
plt.plot([min(true_plot), max(true_plot)],[min(true_plot), max(true_plot)], 'r--')
plt.scatter(true_plot, pred_plot)
plt.show()
```
